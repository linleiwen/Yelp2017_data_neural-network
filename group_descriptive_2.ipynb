```python
import pandas as pd
import numpy as np
import plotly.offline as pyo
from plotly.graph_objs import *
import plotly.plotly as py
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
init_notebook_mode(connected=True)
```

```python
table = pd.read_pickle("dataset/table_whole.pickle")
```

Split table into 2 part (sub1 : those who give 5 star review; sub2 : those who do not give 5 star review)

```python
sub1 = table[table.stars == 5]
```

```python
sub2 = table[table.stars != 5]
```

```python
sub1.describe()
```

```python
sub2.describe()
```

```python
temp = pd.Series([sub1.shape[0],sub2.count()[0]],index = ["5_star_review","non_5_star_review"])
```

```python
temp
```

Pie char of number of 5 star reviews and non 5 star reviews


```python
pieTrace = {'type' : 'pie',
           'labels' : temp.index,
           'values' : temp,
           'marker' : {'colors' : ["rgb(12,192,170)","lightblue"]}
           }

data = [pieTrace]

layout = {'title' : 'Pie char of number of 5 star reviews and non 5 star reviews'}

fig = {'data' : data, 'layout' : layout}

pyo.iplot(fig)
```

```python
data = pd.DataFrame([sub1.describe().ix["mean",:],sub2.describe().ix["mean",:]],index = ["5_star_review","non_5star_review"]).T
```

```python
data
```

```python
data = data.apply(lambda x: x/data["non_5star_review"])
```

```python
data
```

```python
import plotly.offline as pyo
from plotly.graph_objs import *
import plotly.plotly as py
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
init_notebook_mode(connected=True)
```

```python

number= {'type' : 'bar',
                     'x' : data.index,
                     'y' : data["5_star_review"],
                     'marker' : {'color' : 'blue',
                                'line' : {'color' : '#333',
                                          'width' : 2}},
                     'name':"5_star_review",
                     'opacity' : 0.6,}
number2= {'type' : 'bar',
                     'x' : data.index,
                     'y' : data["non_5star_review"],
                     'marker' : {'color' : 'red',
                                'line' : {'color' : '#333',
                                          'width' : 2}},
                      'name':"non_5_star_review",
                     'opacity' : 0.4,}
                     

layout = {'title' : "Percentage difference of characteristics between 5 star review and non 5 year review ",
         'xaxis' : {'title' : 'characteristics'},
         'yaxis' : {'title' : 'percentage'},
         }
fig = {'data' : [number,number2],
      'layout' : layout}

pyo.iplot(fig)
```
